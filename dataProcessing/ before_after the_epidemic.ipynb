```python
df_path = '/content/drive/MyDrive/MACSS60000/FINAL_DATA/control_mental.csv'
```

```python
df_all_data = pd.read_csv(df_path)
df_all_data
```

```python
df_all_data['tokenized_text'] = df_all_data['selftext'].apply(lambda x: lucem_illud.word_tokenize(x))
df_all_data['normalized_tokens'] = df_all_data['tokenized_text'].apply(lambda x: lucem_illud.normalizeTokens(x))
```

```python
df_all_data.head()
```

```python
token_data = '/content/drive/MyDrive/MACSS60000/FINAL_DATA/token_text_data.csv'
```

```python
token_data_df = pd.read_csv(token_data)
token_data_df.head()
```

```python
# Count words in each 'selftext'
def simple_word_count(text):
    return len(str(text).split())

df_all_data['word_count'] = df_all_data['selftext'].apply(simple_word_count)

df_all_data[['selftext', 'word_count']].head()
```

```python
from nltk import ngrams
```

```python
def generate_ngrams(tokens, n=2):
    return list(ngrams(tokens, n))
```

```python
df_all_data['bigrams'] = df_all_data['normalized_tokens'].apply(lambda tokens: generate_ngrams(tokens, n=2))
```

```python
df_all_data['bigrams'].head()
```

```python
from collections import Counter

all_bigrams = sum(df_all_data['bigrams'].tolist(), [])
bigram_freq = Counter(all_bigrams)
bigram_freq.most_common(100)
```

```python
group_0_rows = df_all_data[df_all_data['group'] == 0]
all_bigrams_group_0 = sum(group_0_rows['bigrams'].tolist(), [])
bigram_freq_group_0 = Counter(all_bigrams_group_0)
print(bigram_freq_group_0.most_common(100))
```

```python
group_1_rows = df_all_data[df_all_data['group'] == 1]
all_bigrams_group_1 = sum(group_1_rows['bigrams'].tolist(), [])
bigram_freq_group_1 = Counter(all_bigrams_group_1)
print(bigram_freq_group_1.most_common(100))
```

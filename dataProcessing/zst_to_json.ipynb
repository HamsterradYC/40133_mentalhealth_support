```python
def decompress_zst_to_json_file(zst_file_path, output_file_path):
    with open(zst_file_path, 'rb') as compressed_file:
        decompressor = zstd.ZstdDecompressor()
        with decompressor.stream_reader(compressed_file) as reader:
            decompressed_data = reader.read().decode('utf-8')
            json_objects = decompressed_data.split('\n')
            with open(output_file_path, 'w', encoding='utf-8') as output_file:
                for obj in json_objects:
                    if obj.strip():
                        json_obj = json.loads(obj)
                        output_file.write(json.dumps(json_obj))
                        output_file.write('\n')
```

```python
zst_file_path = '/content/drive/My Drive/MACSS60000/FINAL_DATA/mental_health/zst/mentalillness_comments.zst'
output_file_path = '/content/drive/My Drive/MACSS60000/FINAL_DATA/mental_health/json/mentalillness_comments.json'
decompress_zst_to_json_file(zst_file_path, output_file_path)
print("Decompression and JSON parsing completed. Data written to:", output_file_path)
```

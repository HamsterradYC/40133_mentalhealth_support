```python
import re
df['selftext_sentiment_probabilities'] = df['selftext_sentiment_probabilities'].apply(
    lambda x: [float(i) for i in re.split('\s+', x.strip('[]')) if i]
)
```

```python
import pandas as pd
import numpy as np
from ast import literal_eval
from tqdm import tqdm


def calculate_sentiment_with_threshold(probabilities, threshold=1):
    if any(p >= threshold for p in probabilities):
        max_prob_index = probabilities.index(max(probabilities))
        normalized_sentiment = (max_prob_index - 2) / 2
    else:
        ratings = np.array([1, 2, 3, 4, 5])
        expected_value = np.dot(probabilities, ratings)
        normalized_sentiment = (expected_value - 3) / 2
    return normalized_sentiment


tqdm.pandas(desc="Calculating Sentiments")
df['selftext_normalized_sentiment'] = df['selftext_sentiment_probabilities'].progress_apply(calculate_sentiment_with_threshold)
```
